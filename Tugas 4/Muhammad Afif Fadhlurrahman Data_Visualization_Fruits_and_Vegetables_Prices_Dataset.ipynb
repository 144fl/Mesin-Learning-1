# Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'productname': ['Product1', 'Product2', 'Product3'],
    'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'farmprice': ['5.00', '6.00', '4.50'],
    'atlantaretail': ['7.00', '7.50', '6.80'],
    'chicagoretail': ['6.50', '6.80', '6.30'],
    'losangelesretail': ['8.00', '8.20', '7.90'],
    'newyorkretail': ['7.20', '7.40', '7.00'],
    'averagespread': ['2.00', '1.50', '1.70']
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Mengonversi kolom-kolom ke tipe data numerik
numeric_columns = ['farmprice', 'atlantaretail', 'chicagoretail', 'losangelesretail', 'newyorkretail', 'averagespread']
df[numeric_columns] = df[numeric_columns].astype(float)

# Menampilkan boxplot untuk setiap fitur numerik
plt.figure(figsize=(10, 6))
df.boxplot(column=numeric_columns, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot of Numeric Features')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()

```

#Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'productname': ['Product1', 'Product2', 'Product3'],
    'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'farmprice': [5.00, 6.00, 4.50],
    'atlantaretail': [7.00, 7.50, 6.80],
    'chicagoretail': [6.50, 6.80, 6.30],
    'losangelesretail': [8.00, 8.20, 7.90],
    'newyorkretail': [7.20, 7.40, 7.00],
    'averagespread': [2.00, 1.50, 1.70]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Mengambil hanya kolom-kolom numerik
numeric_columns = df.select_dtypes(include=['float64']).columns

# Menampilkan scatter plot untuk setiap pasangan fitur numerik
plt.figure(figsize=(12, 8))
for i in range(len(numeric_columns)):
    for j in range(i+1, len(numeric_columns)):
        plt.subplot(len(numeric_columns), len(numeric_columns), i * len(numeric_columns) + j - i)
        plt.scatter(df[numeric_columns[i]], df[numeric_columns[j]])
        plt.xlabel(numeric_columns[i])
        plt.ylabel(numeric_columns[j])

plt.tight_layout()
plt.show()
```

#Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    'productname': ['Product1', 'Product2', 'Product3'],
    'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'farmprice': [5.00, 6.00, 4.50],
    'atlantaretail': [7.00, 7.50, 6.80],
    'chicagoretail': [6.50, 6.80, 6.30],
    'losangelesretail': [8.00, 8.20, 7.90],
    'newyorkretail': [7.20, 7.40, 7.00],
    'averagespread': [2.00, 1.50, 1.70]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Mengonversi kolom-kolom ke tipe data numerik
numeric_columns = df.select_dtypes(include=['float64']).columns

# Menghitung korelasi antar fitur numerik
correlation_matrix = df[numeric_columns].corr()

# Membuat heatmap untuk semua fitur numerik
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numerical Features')
plt.show()

```
